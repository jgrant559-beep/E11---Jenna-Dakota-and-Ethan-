```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
str_creek = pd.read_csv("https://raw.githubusercontent.com/jgrant559-beep/E11---Jenna-Dakota-and-Ethan-/refs/heads/main/Week6/strawberry_creek.csv")

str_creek = str_creek[1:].reset_index(drop=True)
```

```python
humidity = str_creek.loc[:, "Humidity"]
temperature = str_creek.loc[:, "Temperature"]

plt.scatter(humidity, temperature, alpha=0.65)
plt.xlabel("Humidity")
plt.ylabel("Temperature")
plt.title("Temperature vs. Humidity")
plt.show()
```

Temperature and Humidity seem to have a correlation. The as humidity goes up, temperature tends to get lower. I could quantify this with a linear regression line. If the calculated R^2 value is close enough to 1, I will become more confident in the the relationship between these two variables.

```python
def regression(x, y):
    m, b = np.polyfit(x, y, 1)
    xs = np.linspace(np.min(x)-1, np.max(x)+1, 100)
    
    plt.scatter(x, y, alpha=0.65)
    plt.plot(xs, m*xs + b, color="red")
    plt.xlabel(f"{x.name}")
    plt.ylabel(f"{y.name}")
    plt.title(f"{y.name} vs. {x.name}")
    plt.show()
    
    y_bar = np.mean(y)
    sstot = np.sum((y-y_bar)**2)
    ssres = np.sum((y - (m*x + b))**2)
    print(f"We have an R^2 value of: {1 - ssres/sstot}")
regression(humidity, temperature)
```

```python
# display(str_creek)
column_names = ["Particles > 0.3um / 0.1L air", "Gas", "Pressure", "Altitude"]
for var in column_names:
    data = str_creek.loc[:, var]
    regression(data, temperature)
```

The relationships between temperature and the other variables seem to have a lot weaker correlation than the relationship between temperature and humidity.
