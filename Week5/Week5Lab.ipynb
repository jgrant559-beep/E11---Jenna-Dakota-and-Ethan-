```python
import pandas as pd
import matplotlib.pyplot as plt
import math
rawInsideData = pd.read_csv('inside_data.csv')
rawOutsideData = pd.read_csv('outside_data.csv')

insideData = rawInsideData[1:].reset_index(drop=True)
outsideData = rawOutsideData[60:].reset_index(drop=True)
```

```python
#Plots 
column_names = rawInsideData.columns

for i in range(6): 
    currentLabel = column_names[i+1]
    
    plt.plot(outsideData.iloc[: , i+1], label="Outside", alpha = .7, color = "red")
    plt.plot(insideData.iloc[:, i+1], label="Inside", alpha = .7, color = "blue")
    plt.ylabel(currentLabel)
    plt.xlabel("Seconds")
    plt.title(currentLabel + " vs Seconds") 
    plt.legend()
    plt.show()
```

```python
#Histograms & Error Analysis

for i in range(6): 
    currentLabel = column_names[i+1]
    currentOutside = outsideData.iloc[: , i+1]
    currentInside = insideData.iloc[: , i+1]
    
    meansOutside = currentOutside.mean()
    meansInside = currentInside.mean()
    
    stdOutside = currentOutside.std()
    stdInside = currentInside.std()

    standardErr = math.sqrt((stdInside ** 2 / (currentInside.size))+(stdOutside ** 2 / (currentOutside.size)))

    print("Difference in means is > 3 : " + str(abs(meansOutside-meansInside) > (3 * standardErr)))
        
    
    plt.hist(currentOutside, label="Outside", alpha = .7, color = "red")
    plt.hist(currentInside, label="Inside", alpha = .7, color = "blue")
    plt.xlabel(currentLabel)
    plt.ylabel("Frequency") 
    plt.legend()
    plt.show()
```
